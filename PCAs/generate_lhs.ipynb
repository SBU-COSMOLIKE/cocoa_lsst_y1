```python
import numpy as np
from scipy.stats.qmc import LatinHypercube as LHS, scale
```

```python
full_limits = np.array([
    [0.24, 0.4],      # Omegam
    [0.04, 0.06],     # Omegab
    [0.92, 1.0],      # ns
    [1.7e-9, 2.5e-9], # As
    [0.61, 0.73],     # h
])
```

```python
lengths = (full_limits[:, 1] - full_limits[:, 0])
```

```python
new_mins = (full_limits[:, 0] + 0.15*lengths)
new_maxs = (full_limits[:, 1] - 0.15*lengths)
new_lims = np.dstack((new_mins, new_maxs))
```

```python
sampler = LHS(d=5)
samples = scale(sampler.random(n=30), new_mins, new_maxs)
```
