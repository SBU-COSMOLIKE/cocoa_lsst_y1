# COLA Emulator - Example
The usage of COLAEMU is almost identical to Euclid Emulator 2.

```python
import matplotlib.pyplot as plt
import nn_emu_lcdm as colaemu
import nn_emu_wcdm as colaemu_wcdm
import euclidemu2 as ee2
```

```python
ks_cola = colaemu.cola_ks_default
```

```python
# Testing LCDM cosmology
params = {
    'Omm': 0.319,
    'Omb': 0.049,
    'As': 2.1e-9,
    'ns': 0.96,
    'h': 0.67
}
_, boost = ee2.get_boost(params, colaemu.cola_redshifts, ks_cola)
boost_cola = colaemu.get_boost(params)
```

```python
z_index = 0
plt.semilogx(ks_cola, boost[z_index], label="EE2")
plt.semilogx(ks_cola, boost_cola[z_index], label="COLAEMU")
plt.title(f"Nonlinear Boosts for EE2 Reference Cosmology at z = {colaemu.cola_redshifts[z_index]:.2f}")
plt.xlabel("k")
plt.ylabel("Boost")
plt.legend()
plt.show()

plt.semilogx(ks_cola, boost_cola[z_index]/boost[z_index] - 1)
plt.xlabel("k")
plt.ylabel("Fractional Difference")
plt.title(f"Fractional Difference between COLAEMU and EE2 at z = {colaemu.cola_redshifts[z_index]:.2f}")
```

```python
z_index = 34
plt.loglog(ks_cola, boost[z_index], label="EE2")
#plt.loglog(ks_cola, boost_cola[z_index], label="COLAEMU")
plt.xlim([1, 3])
plt.axvline(3.1416/2)
plt.ylim([2, 3])
plt.title(f"Nonlinear Boosts for EE2 Reference Cosmology at z = {colaemu.cola_redshifts[z_index]:.2f}")
plt.xlabel("k")
plt.ylabel("Boost")
plt.legend()
plt.show()
```
